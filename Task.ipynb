```python
import os
import openai

# ⬇️ Replace ONLY after generating a NEW key
os.environ["OPENAI_API_KEY"] = "Your Open AI Key"

client = openai.OpenAI(api_key=os.environ["OPENAI_API_KEY"])
```

```python
import textwrap
import json

metadata = {
    "title": "Ramayana",
    "themes": [
        "dharma",
        "loyalty",
        "exile",
        "good vs evil",
        "responsible leadership"
    ],
    "characters": [
        "Rama", "Sita", "Lakshman", "Bharat", "Ravana", "Hanuman"
    ],
    "key_scenes": [
        "Exile of Rama",
        "Life in the forest",
        "Abduction of Sita",
        "Alliance with Hanuman",
        "Siege of Lanka & defeat of Ravana"
    ],
}

setting = "interplanetary cyberpunk rebellion in a far-future galaxy"

def llm(prompt, tokens=1000, temperature=0.8):
    response = client.chat.completions.create(
        model="gpt-5.1",
        messages=[{"role": "user", "content": prompt}],
        max_tokens=tokens,
        temperature=temperature,
    )
    return response.choices[0].message.content
```

```python
def prompt_world():
    return f"""
You are a mythic science fiction world-builder.

Reimagine the Ramayana in this setting:
→ {setting}

Write 600–800 words describing:
- planets, factions, tech, social structure
- ethics as Dharma in post-AI civilization
- tension between tradition & transhumanism
- tone: poetic sci-fi, but coherent

NO bullets. Output full prose.
"""

def prompt_characters(world):
    return f"""
You are a narrative systems designer.

Based on this world:
{world[:700]}

Task:
1. Create a table mapping original characters to new versions
2. Include:
   - new name
   - new role
   - evolved motivation
3. Write 2 short paragraphs explaining how relationships change

Output:
- Markdown table
- Then paragraphs
"""

def prompt_plot(world, chars):
    return f"""
Write a 3-act plot outline for the reimagined Ramayana.

Must include:
- cyberpunk exile sequence
- high-tech Sita abduction
- planetary siege equivalent
- thematic Dharmic resolution

Take the original key scenes and reinterpret them.

Length: ~700 words.
"""

def prompt_reinterpret(plot):
    return f"""
Analyze this reinterpretation.

Include:
1. Core dramatic question (original & new)
2. 2–3 parallel scenes & why they work

Use section headings.
"""

def prompt_rationale(world, plot):
    return f"""
Write a 400–500 word creative rationale explaining:
- why this reinterpretation works
- how emotion & theme are preserved
- how sci-fi deepens the meaning
- what is gained by choosing cyberpunk

Tone: clear, intelligent, reflective.
"""
```

```python
print("Generating world...")
world = llm(prompt_world())

print("\nGenerating character mapping...")
characters = llm(prompt_characters(world))

print("\nGenerating plot...")
plot = llm(prompt_plot(world, characters))

print("\nGenerating plot reinterpretation...")
reinterpretation = llm(prompt_reinterpret(plot))

print("\nGenerating rationale...")
rationale = llm(prompt_rationale(world, plot))
```

```python
final = f"""
# Reimagined Story: Ramayana (Cyberpunk Space Rebellion)

## 1. Worldbuilding
{world}

## 2. Character Mapping & Relationships
{characters}

## 3. Plot (Acts & Beats)
{plot}

## 4. Plot Reinterpretation
{reinterpretation}

## 5. Creative Rationale
{rationale}
"""

print(final[:2500])  # preview first part

with open("ramayana_output.md", "w") as f:
    f.write(final)

print("\n\n✔️ Saved as ramayana_output.md")
```
